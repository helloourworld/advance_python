```python
# read the raw data
def readData(filename,start_date = pd.datetime.today().date() - datetime.timedelta(90)
,end_date=pd.datetime.today().date()):
    data = pd.read_csv(filename+".csv")
    dt = data["date"].apply(lambda x: pd.to_datetime(x).date())  # turn to date
    data.set_index(dt, inplace=True, drop=True)# set index
    data.sort_index(ascending=True, inplace=True)
    data = data[(data.index >= pd.to_datetime(start_date).date()) & (data.index <  pd.to_datetime(end_date).date())] # 选择时间范围
    data = data[['price']]
    data.columns = [filename]
    return data

# cal the cum ret
def calCumRet(secList, start_date = pd.datetime.today().date() - datetime.timedelta(90), end_date=pd.datetime.today().date()):
    da = []
    for each in secList:
        print each
        if each in ['hs300','sz50']:
            data = ts.get_hist_data(each, str(start_date), str(end_date))
            dt = pd.Series(data.index.values).apply(lambda x: pd.to_datetime(x).date())
            data.set_index(dt, inplace=True, drop=True)# set index
            data.sort_index(ascending=True, inplace=True)
            da += [data['close']]
        else:
            hist_data_each = readData(each, start_date, end_date)
            da += [hist_data_each]
    data = pd.concat(da, axis=1)
    data.columns = secList
    data = data.dropna(axis = 0, how = 'any')
    cumRetdata = data.apply(lambda df: df /df [0]) -1 # 按照基准为%0计算超额收益
    cumRetdata = cumRetdata.dropna(axis = 0, how = 'any')
    return data, cumRetdata
```

```python
def cons_Portfolio(cumRetData):
    n = 1
    Portfolio = []
    portRetData = []
    for i in range(1, 11):
        for j in range(1,4):
            port = dfc[(dfc['term'] == j) & (dfc['risk'] == i)]
            # print port.head(1)
            
            Portfolio.append(port)
            portRet = pd.DataFrame(np.mat(cumRetData[port['code']]) * np.mat(port[['weight']]))
            portRet.set_index(cumRetData.index,  inplace=True, drop=True)# set index
            portRet.columns = ["Portfolio" + str(n)]
            portfolioData = pd.concat([portRet, benchRetCumData], axis=1).dropna(axis=0,how='any')
            # portfolioData.reset_index(drop=True, inplace=True)
            portRetData.append(portfolioData)
            
            n += 1
    return portRetData
```

### 3.1 大类资产权重MC 模拟

我们最想知道的是给定的一个股票池（证券组合）如何找到风险和收益平衡的位置。
下面通过一次蒙特卡洛模拟，产生大量随机的权重向量，并记录随机组合的预期收益和方差。

```python
def get_annual_pct_ret(data):
    return data.pct_change().dropna()

def get_annual_log_ret(data):
    return np.log(data.diff()).dropna()

def get_covmat(retData, tradeDay=252):
    return retData.cov()*tradeDay

def plot_capm(start_dt, end_dt, dropn = 1, mc = 10000, Rf = 0.04):
    port_returns = []
    port_variance = []
    data = calCumRet(secList, start_dt, end_dt)[0]
    # retData = data.pct_change().dropna()
    retData = np.log(data / data.shift(1)).dropna()
    
    retData = retData[secList[:-dropn]] # drop hs300 217004BI
    # return retData
    # print cov_mat
    cov_mat = get_covmat(retData)*100
    print cov_mat
    noa = cov_mat.shape[0]
    for p in range(mc):
        weights = np.random.random(noa)
        weights /= np.sum(weights)
        port_returns.append(np.sum(retData.mean()*252*weights))
        port_variance.append(np.sqrt(np.dot(weights.T, np.dot(retData.cov()*252, weights))))
    
    port_returns = np.array(port_returns)
    port_variance = np.array(port_variance)

    #无风险利率设定为4%
    risk_free = 0.0443
    plt.figure(figsize = (10,5))
    plt.scatter(port_variance, port_returns, c=(port_returns-risk_free)/port_variance, marker = 'o')
    plt.grid(True)
    plt.xlabel('excepted volatility')
    plt.ylabel('expected return')
    plt.colorbar(label = 'Sharpe ratio')
    plt.title(start_dt + "--" + end_dt + " MC Simulation")

def plot_retData(start_dt, end_dt, dropn = 1, mc = 10000, Rf = 0.04):
    port_returns = []
    port_variance = []
    data = calCumRet(secList, start_dt, end_dt)[0]
    # retData = data.pct_change().dropna()
    retData = np.log(data / data.shift(1)).dropna()
    
    retData = retData[secList[:-dropn]] # drop hs300 217004BI
    return retData
```


### 3.2 大类资产权重最优化模拟

#### 3.2.1 SharpeRatio 最大

建立statistics函数来记录重要的投资组合统计数据（收益，方差和夏普比）

通过对约束最优问题的求解，得到最优解。其中约束是权重总和为1。

```python
def get_weights(retData):
    
    cov_mat = get_covmat(retData)
    # 根据上面的描述写好对应的P、q、G、h、A、b
    Nums = cov_mat.shape[0]
    P = matrix(cov_mat.values)
    q = matrix(np.zeros(Nums))
    G = matrix(np.diag(-np.ones(Nums)))
    h = matrix(np.zeros(Nums))
    A = matrix(np.ones(Nums)).T
    b = matrix(1.0).T

    # 代入优化方程求解
    solvers.options['show_progress'] = False
    sol = solvers.qp(P, q, G, h, A, b)
    wts = pd.Series(index=cov_mat.index, data=np.array(sol['x']).flatten())
    wts = wts[wts >= 0.01]  # weights调整，对于权重太小的不予配置
    wts_adjusted = wts / wts.sum() * 1.0
    return wts_adjusted
```

```python
def statistics(weights):
    weights = np.array(weights)
    port_returns = np.sum(retData.mean()*weights)*252
    port_variance = np.sqrt(np.dot(weights.T, np.dot(retData.cov()*252,weights)))
    return np.array([port_returns, port_variance, port_returns/port_variance])

#最小化夏普指数的负值
def min_sharpe(weights):
    return -statistics(weights)[2] 

#最优化投资组合的推导是一个约束最优化问题
import scipy.optimize as sco
def calOptSharpe():
    #约束是所有参数(权重)的总和为1。这可以用minimize函数的约定表达如下
    cons = ({'type':'eq', 'fun':lambda x: np.sum(x)-1})
   
    #我们还将参数值(权重)限制在0和1之间。这些值以多个元组组成的一个元组形式提供给最小化函数
    bnds = tuple((0,1) for x in range(noa))
    
    #优化函数调用中忽略的唯一输入是起始参数列表(对权重的初始猜测)。我们简单的使用平均分布。
    opts = sco.minimize(min_sharpe, noa*[1./noa,], method = 'SLSQP', bounds = bnds, constraints = cons)
    return opts
```

```python
# retData = get_annual_pct_ret(benchmarkdata)
noa = retData.shape[1]
opts = calOptSharpe()
```

### 3.2.2 投资组合优化2——方差最小

接下来，我们通过方差最小来选出最优投资组合。

```python
#但是我们定义一个函数对 方差进行最小化
def min_variance(weights):
    return statistics(weights)[1]
cons = ({'type':'eq', 'fun':lambda x: np.sum(x)-1})
bnds = tuple((0,1) for x in range(noa))
optv = sco.minimize(min_variance, noa*[1./noa,],method = 'SLSQP', bounds = bnds, constraints = cons)
```

#### 3.2.3 组合的有效前沿

有效前沿有既定的目标收益率下方差最小的投资组合构成。

在最优化时采用两个约束，1.给定目标收益率，2.投资组合权重和为1。

```python
def min_variance(weights):
    return statistics(weights)[1]

#在不同目标收益率水平（target_returns）循环时，最小化的一个约束条件会变化。
target_returns = np.linspace(0.0,0.5,50)
target_variance = []
for tar in target_returns:
    cons = ({'type':'eq','fun':lambda x:statistics(x)[0]-tar},{'type':'eq','fun':lambda x:np.sum(x)-1})
    res = sco.minimize(min_variance, noa*[1./noa,],method = 'SLSQP', bounds = bnds, constraints = cons)
    target_variance.append(res['fun'])

target_variance = np.array(target_variance)
```

下面是最优化结果的展示。

**叉号**：构成的曲线是有效前沿（目标收益率下最优的投资组合）

**红星**：sharpe最大的投资组合

**黄星**：方差最小的投资组合

```python
def plot_Opt(start_dt, end_dt, dropn = 1, mc = 10000, Rf = 0.04):
    port_returns = []
    port_variance = []
    data = calCumRet(secList, start_dt, end_dt)[0]
    retData = data.pct_change().dropna()
    retData = retData[secList[:-dropn]] # drop hs300 217004BI
    # print cov_mat
    cov_mat = get_covmat(retData)*100
    print cov_mat
    noa = cov_mat.shape[0]
    for p in range(mc):
        weights = np.random.random(noa)
        weights /= np.sum(weights)
        port_returns.append(np.sum(retData.mean()*252*weights))
        port_variance.append(np.sqrt(np.dot(weights.T, np.dot(retData.cov()*252, weights))))
    
    port_returns = np.array(port_returns)
    port_variance = np.array(port_variance)

    plt.figure(figsize = (8,4))
    #圆圈：蒙特卡洛随机产生的组合分布
    plt.scatter(port_variance, port_returns, c = port_returns/port_variance,marker = 'o')
    #叉号：有效前沿
    plt.scatter(target_variance,target_returns, c = target_returns/target_variance, marker = 'x')
    #红星：标记最高sharpe组合
    plt.plot(statistics(opts['x'])[1], statistics(opts['x'])[0], 'r*', markersize = 15.0)
    #黄星：标记最小方差组合
    plt.plot(statistics(optv['x'])[1], statistics(optv['x'])[0], 'y*', markersize = 15.0)
    plt.grid(True)
    plt.xlabel('expected volatility')
    plt.ylabel('expected return')
    plt.colorbar(label = 'Sharpe ratio')
```

###  3.3 大类资产权重输出

```python
def target_weights(returns, noa, risk, benchmark_returns=None):
    def statistics(weights):
        '''
        计算 收益，波动率，p
        :param weights:
        :return:
        '''
        weights = np.array(weights)
        port_returns = np.sum(returns.mean() * weights) * 252
        port_variance = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
        p = stats.norm.cdf(-risk, 0, np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights))))
        return np.array([port_returns, port_variance, port_returns / port_variance, p])

    def min_returns(weights):
        return -statistics(weights)[0]

    def min_beta(weights):
        cov = np.cov(np.sum(weights * returns, axis=1), benchmark_returns)
        return - cov[0][1] / cov[1][1]

    # bnds = tuple((0.15, 1) for x in range(noa))
    bnds = tuple((0.05*min((i+1),3), 1) for i,x in enumerate(range(noa)))

    if isinstance(benchmark_returns, pd.DataFrame):
        cons = ({'type': 'ineq', 'fun': lambda x: risk - statistics(x)[3]},
                {'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
        res = sco.minimize(min_beta, noa * [1. / noa, ],
                           method='SLSQP',
                           bounds=bnds,
                           constraints=cons)
    else:
        cons = ({'type': 'eq', 'fun': lambda x: statistics(x)[1] - risk},
                {'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

        res = sco.minimize(min_returns, noa * [1. / noa, ],
                           method='SLSQP',
                           bounds=bnds,
                           constraints=cons)
    return res['x']
```

```python
def OptWeight(risk):
    # risk = np.array(risk)
    weights_Opt = []
    for each in risk:
        weight_t = target_weights(retData, noa, each, benchmark_returns=None)
        weights_Opt.append(weight_t)
    return weights_Opt
```

```python
from sklearn import linear_model
```

```python
# Function to get data
def get_data(data):

    X_parameter = []
    Y_parameter = []
    for single_square_feet ,single_price_value in zip(data['vol'],data['ret']):
        X_parameter.append([float(single_square_feet)])
        Y_parameter.append(float(single_price_value))
    return X_parameter,Y_parameter

```

```python
# Function for Fitting our data to Linear model
def linear_model_main(X_parameters,Y_parameters,predict_value):
    # Create linear regression object
    regr = linear_model.LinearRegression()
    regr.fit(X_parameters, Y_parameters)
    predict_outcome = regr.predict(predict_value)
    predictions = {}
    predictions['intercept'] = regr.intercept_
    predictions['coefficient'] = regr.coef_
    predictions['predicted_value'] = predict_outcome
    return predictions
```

```python
X,Y = get_data(dfa)
```

```python
predictvalue = X[0]
result = linear_model_main(X,Y,predictvalue)
print "Intercept value " , result['intercept']
print "coefficient" , result['coefficient']
print "Predicted value: ",result['predicted_value']
```

```python
# Function to show the resutls of linear fit model
def show_linear_line(X_parameters,Y_parameters):
    # Create linear regression object
    regr = linear_model.LinearRegression()
    regr.fit(X_parameters, Y_parameters)
    plt.rcParams['font.sans-serif'] = ['SimHei']# 用来正常显示中文标签
    plt.scatter(X_parameters,Y_parameters,color='blue')
    plt.plot(X_parameters,regr.predict(X_parameters),color='red',linewidth=4)
    plt.title(u"图2 预期收益与预期风险回归分析")
    plt.xlabel(u"预期风险")
    plt.ylabel(u"预期收益")
    plt.show()
```
